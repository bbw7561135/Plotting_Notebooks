```python
filename = 'radave300.txt'
X1,Y1 = [],[]
with open(filename,'r') as f:
    lines = f.readlines()
    for line in lines:
        value = [float(s) for s in line.split()]
        X1.append(value[0])
        Y1.append(value[1])
X1new = np.array(X1)
Y1new = np.array(Y1) # in order for mean() and other statistics funs
#print(X1new)
#print(Y1new)
#print(Y1new.mean())



f2 = plt.figure(figsize=[12,8]) 
ax2 = f2.add_subplot(111) 
#plt.plot(Xnew,Ynew)
plt.plot(X1new,Y1new,label='shock radius at 1000 years')
#plt.plot(X1new,Y1new+0.05,label='counting area')
#plt.hlines(y=Y1new.mean(),xmin=0.0,xmax=180.0,colors='r',linestyles='--',label='averaged shock radius')
plt.fill_between(X1new,Y1new,Y1new+0.025,facecolor='grey',alpha=0.3,label='counting region')
plt.fill_between(X1new,Y1new-0.025,Y1new,facecolor='grey',alpha=0.3)
plt.grid(True)
plt.minorticks_on()
plt.legend(loc=0,ncol=1)#0 is auto-set while ncol is lie shu 
font1={'size':15}
ax2.set_xlabel(r'Azimuth Angles(deg)',fontdict=font1) 
ax2.set_ylabel(r'Radius(pc)',fontdict=font1) 
ax2.axis([0.0,180.0,5.0,7.5])
#ax1.set_title(r'radius profile with t=1000 years')
plt.savefig('radius_average_2048.png',bbox_inches='tight')#不能存为pdf会跑挂
```
