```python
filename = 'radave300.txt'
X4,Y4 = [],[]
with open(filename,'r') as f:
    lines = f.readlines()
    for line in lines:
        value = [float(s) for s in line.split()]
        X4.append(value[0])
        Y4.append(value[1])
X4new = np.array(X4)
Y4new = np.array(Y4) # in order for mean() and other statistics funs
#print(X4new)
#print(Y4new)
#print(Y4new.mean())



f3 = plt.figure(figsize=[9,6]) 
ax1 = f3.add_subplot(111) 
#plt.plot(Xnew,Ynew)
plt.plot(X4new,Y4new,label='shock radius')
plt.hlines(y=Y4new.mean(),xmin=0.0,xmax=90.0,colors='r',linestyles='--',label='averaged shock radius')
plt.grid(True)
plt.minorticks_on()
plt.legend(loc=0,ncol=1)#0 is auto-set while ncol is lie shu 
font1={'size':12}
ax1.set_xlabel(r'Angles(deg)',fontdict=font1) 
ax1.set_ylabel(r'Radius(pc)',fontdict=font1) 
ax1.axis([0.0,90.0,6.5,7.05])
#ax1.set_title(r'radius profile with t=1000 years')
plt.savefig('radius_average.png')#pdf
```
